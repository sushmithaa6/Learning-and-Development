```python
#Matplotlib

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Demo graph')
```

```python
y=x*x
plt.plot(x,y,'ro')
```

```python
plt.plot(x,y,'c*--')
```

```python
plt.plot(x,y,'b*',linestyle='dotted',linewidth=4,markersize=10)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Squares')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'b--')
plt.subplot(2,2,3)
plt.plot(x,y,'go')
plt.subplot(2,2,4)
plt.plot(x,y,'y*')
```

```python
x=np.arange(1,11)
y=5*x+7
plt.plot(x,y,'*')
```

```python
#Sub plots

x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y,'r')
plt.title('Sine wave')
```

```python
x=np.arange(0,5*np.pi,0.1)
plt.subplot(1,2,1)
plt.plot(x,np.sin(x),'r--')
plt.subplot(1,2,2)
plt.plot(x,np.cos(x),'g--')
```

```python
#Bar graphs

x=[2,4,6,8]
y=[11,16,9,5]
x1=[3,5,7,9]
y1=[10,15,8,4]
plt.bar(x,y,align='center',color='c')
plt.bar(x1,y1,color='b')

plt.title("Bar graph")
```

```python
#Histogram

a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,color='g')
plt.title('Histogram')
```

```python
#Box plot

d=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(d,vert=True,patch_artist=True);  
```

```python
#Pie chart

labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['lightgreen','yellow','lightcoral','skyblue']
explode=(0.1,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
