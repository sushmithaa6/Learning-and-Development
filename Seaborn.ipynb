```python
#Seaborn

import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.dtypes
```

```python
#Pearson correlation - ranges between 0 and 1

df1=df.drop(['sex','smoker','day','time'], axis=1)
df1.corr()
```

```python
sns.heatmap(df1.corr(),cmap='copper',annot=True,linewidth=1)
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex',color='lightblue')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg',color='cyan')
```

```python
#Scatter plot

sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.histplot(df['tip'],kde=True)
```

```python
sns.histplot(df['tip'],kde=False,bins=10)
#KDE - Kernel Density Estimation
```

```python
#Count plot

sns.countplot(x='day',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
#Bar plot

sns.barplot(x='total_bill',y='sex',data=df)
```

```python
#Box plot

sns.boxplot(y='total_bill',x='sex',data=df)
```

```python
sns.boxplot(x="day",y="total_bill",hue='day',data=df,legend=False)
```

```python
sns.boxplot(data=df,orient='v',palette='rainbow')
```

```python
sns.boxplot(x="total_bill",y="day",hue="smoker",data=df)
```

```python
#Violin plot

sns.violinplot(x="total_bill",y="day",hue='day',data=df,palette='rainbow')
```
